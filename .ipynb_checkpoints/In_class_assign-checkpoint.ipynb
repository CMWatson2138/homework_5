```python
def sort_func (number):
    if type(number)==int and number >= 0:
        numberlist=[]
        for i in str(number):
            numberlist.append(i)
        strlist=[int(i) for i in numberlist]
        sortlist=sorted(strlist, reverse=True)
        number=""
        for i in sortlist:
            number="".join([str(i) for i in sortlist])
        print(number)
    else:
        print(111111)
        
sort_func(9591)        
```

```python
import pandas as pd
import numpy as np

file_to_load=(r'C:\Users\watso\Documents\Data_Science_Bootcamp\homework_5\jeopardy_csv.csv')

jeopardy_data=pd.read_csv(file_to_load)
jeopardy_df=pd.DataFrame(jeopardy_data)
jeopardy_df
```

```python
#There are 216930 rows in this dataframe
#Each row represents information about a question asked in the game.
```

```python
#rename columns to eliminate spaces
jeopardy_df.rename(columns={"Show Number": "Show_Number"}, inplace=True)
jeopardy_df.rename(columns={" Air Date": "Air_Date"}, inplace=True)
#change date information to date time format
jeopardy_df["Air Date"]=pd.to_datetime(jeopardy_df["Air_Date"])
#create a column for the year aired
jeopardy_df['Year_Aired'] = pd.DatetimeIndex(jeopardy_df['Air_Date']).year
#group by year aired
shows_per_year=jeopardy_df.groupby(["Year_Aired"])
#count unique shows by number
unique_shows=shows_per_year.aggregate({"Show_Number": "nunique"})
#rename column
unique_shows.rename(columns={"Show_Number": "Unique_Shows_per_Year"}, inplace=True)
display(unique_shows)
```

```python
#What are the top five categories on the show?
#Answer: Before & After, Science, Literature, American History, and Potpourri

#First group data by category
most_pop_df=jeopardy_df.groupby(" Category")
#Next count number of times a category appears
most_popular=most_pop_df.aggregate({" Category":"count"})
#rename column
most_popular.rename(columns={" Category": "Number_of_Questions"}, inplace=True)
#sort by descending order
most_popular_sorted=most_popular.sort_values("Number_of_Questions", ascending=False)

display(most_popular_sorted.head(5))
```
